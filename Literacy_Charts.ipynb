```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#

literacy_all_csv = 'Resources/literacy_ratios/litall_ratio.csv'
literacy_all = pd.read_csv(literacy_all_csv)
```

```python
literacy_all = literacy_all.set_index("Country Code")

#literacy_all
```

```python
# literacy_data = literacy_data.set_index("Country Code")
# #literacy_data
```

```python
years = literacy_all.keys()
years
```

```python
#Region
one = literacy_all.loc['EAS', years]
two = literacy_all.loc['ECS', years]
three= literacy_all.loc['LCN', years]
four= literacy_all.loc['MEA', years]
five= literacy_all.loc['SAS', years]
six= literacy_all.loc['SSF', years]
world= literacy_all.loc['WLD', years]

#Income
lower = literacy_all.loc['LIC', years]
lower_mid = literacy_all.loc['LMC', years]
mid_mid= literacy_all.loc['MIC', years]
upper_mid= literacy_all.loc['UMC', years]
EU_high = literacy_all.loc['EUU', years]

#Other groupings
FCS = literacy_all.loc['FCS', years]
HPC = literacy_all.loc['HPC', years]
LDC = literacy_all.loc['LDC', years]
World = literacy_all.loc['WLD', years]
```

```python
#By Region - no values for North America
plt.plot(years, one, marker='', color='#eee0e0', linewidth=2)
plt.plot(years, two, marker='', color='#db9035', linewidth=2)
plt.plot(years, three, marker='', color='#c3cb71', linewidth=2)
plt.plot(years, four, marker='', color='#215b63', linewidth=2)
plt.plot(years, five, marker='', color='#A569BD', linewidth=2)
plt.plot(years, six, marker='', color='#005b96', linewidth=2)
plt.legend()


# markersize=4
# markerfacecolor='blue'
#, linestyle='dashed'

plt.xlabel('Year')
plt.ylabel('Female/Male Literacy Ratio')
plt.title("Litaracy Ratio by Region")
plt.show()
```

```python
#By Income - EU in place of high
plt.plot(years, lower, marker='', color='#eee0e0', linewidth=2)
plt.plot(years, lower_mid, marker='', color='#db9035', linewidth=2)
plt.plot(years, mid_mid, marker='', color='#c3cb71', linewidth=2)
plt.plot(years, upper_mid, marker='', color='#215b63', linewidth=2)
plt.plot(years, EU_high, marker='', color='#A569BD', linewidth=2)
plt.legend()

plt.xlabel('Year')
plt.ylabel('Female/Male Literacy Ratio')
plt.title("Litaracy Ratio by Income Group")
plt.show()

```

```python
plt.plot(years, FCS, marker='', color='#eee0e0', linewidth=2)
plt.plot(years, HPC, marker='', color='#db9035', linewidth=2)
plt.plot(years, LDC, marker='', color='#c3cb71', linewidth=2)
plt.plot(years, World, marker='', color='#215b63', linewidth=2, linestyle='dashed')
plt.legend()
         
plt.xlabel('Year')
plt.ylabel('Female/Male Literacy Ratio')
plt.title("Litaracy Ratio - Heavily Indebted Countries")
plt.show()
```
