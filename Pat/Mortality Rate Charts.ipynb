```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
mort_all_csv = 'Resources/mortdata_groups.csv'
mort_all = pd.read_csv(mort_all_csv)
```

```python
mort_all = mort_all.set_index("Country Code")
#literacy_all
```

```python
mort_all = mort_all[['1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017']] * 1
```

```python
years = mort_all.keys()
years
```

```python
#Region
one = mort_all.loc['EAS', years]
two = mort_all.loc['ECS', years]
three= mort_all.loc['LCN', years]
four= mort_all.loc['MEA', years]
five= mort_all.loc['SAS', years]
six= mort_all.loc['SSF', years]
world= mort_all.loc['WLD', years]

#Income
lower = mort_all.loc['LIC', years]
lower_mid = mort_all.loc['LMC', years]
mid_mid= mort_all.loc['MIC', years]
upper_mid= mort_all.loc['UMC', years]
EU_high = mort_all.loc['EUU', years]

#Other groupings
FCS = mort_all.loc['FCS', years]
HPC = mort_all.loc['HPC', years]
LDC = mort_all.loc['LDC', years]
World = mort_all.loc['WLD', years]
```

```python
#By Region - no values for North America
plt.plot(years, one, marker='', color='#eee0e0', linewidth=2)
plt.plot(years, two, marker='', color='#db9035', linewidth=2)
plt.plot(years, three, marker='', color='#c3cb71', linewidth=2)
plt.plot(years, four, marker='', color='#215b63', linewidth=2)
plt.plot(years, five, marker='', color='#A569BD', linewidth=2)
plt.plot(years, six, marker='', color='#005b96', linewidth=2)
plt.legend()


# markersize=4
# markerfacecolor='blue'
#linestyle='dashed'

plt.xlabel('Year')
plt.ylabel('Mortality Rate')
plt.title("Mortality Rate Under 5 per 1000 Births by Region")
plt.show()
```

```python
#By Income - EU in place of high
plt.plot(years, lower, marker='', color='#eee0e0', linewidth=2)
plt.plot(years, lower_mid, marker='', color='#db9035', linewidth=2)
plt.plot(years, mid_mid, marker='', color='#c3cb71', linewidth=2)
plt.plot(years, upper_mid, marker='', color='#215b63', linewidth=2)
plt.plot(years, EU_high, marker='', color='#A569BD', linewidth=2)
plt.legend()

plt.xlabel('Year')
plt.ylabel('Mortality Rate')
plt.title("Mortality Rate by Income Group")
plt.show()

```

```python
plt.plot(years, FCS, marker='', color='#eee0e0', linewidth=2)
plt.plot(years, HPC, marker='', color='#db9035', linewidth=2)
plt.plot(years, LDC, marker='', color='#c3cb71', linewidth=2)
plt.plot(years, World, marker='', color='#215b63', linewidth=2, linestyle='dashed')
plt.legend()
         
plt.xlabel('Year')
plt.ylabel('Mortality Rate')
plt.title("Mortality Rate - Heavily Indebted Countries")
plt.show()
```
