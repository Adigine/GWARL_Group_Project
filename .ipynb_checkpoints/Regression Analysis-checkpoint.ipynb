```python
x = merge_final['2017_x']
y = merge_final['Ratio of Female to Male Literacy Rates']

# Find the slope and intercept of the best fit line
slope, intercept = np.polyfit(x, y, 1)

# Create a list of values in the best fit line
abline_values = [slope * i + intercept for i in x]

# Plot the best fit line over the actual values
plt.figure(figsize=(10,10))
plt.plot(x, y, 'o')
plt.plot(x, abline_values, 'b')
plt.xlabel('Mortality Rate',size=15)
plt.ylabel('Literacy Rate',size=15)
plt.title('Literacy vs Mortality Rate in 2017',fontweight='bold',size=20)
plt.tight_layout()
#plt.savefig('correlation_lit.png')
plt.show()
```
