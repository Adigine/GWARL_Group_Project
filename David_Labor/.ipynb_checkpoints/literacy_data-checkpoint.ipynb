```python
import pandas as pd
import numpy as np
import matplotlib as plt
import geopandas as gpd
from descartes import PolygonPatch

csv_fem_lit = "../Resources/female_literacy.csv"
csv_male_lit = "../Resources/male_literacy.csv"
# csv_total_lab = "../Resources/labor_total.csv"
# csv_lab_perc_fem = "../Resources/labor_perc_fem.csv"


# Literacy Rate
fem_lit = pd.read_csv(csv_fem_lit)
male_lit = pd.read_csv(csv_male_lit)
```

```python
# Function to clean up data sets and use only country-level data, no regions or WB categories
country_list = ['AFG', 'ALB', 'DZA', 'ASM', 'AND', 'AGO', 'AIA', 'ATG', 'ARG', 'ARM', 'ABW', 'AUS', 'AUT', 'AZE', 'BHS', 'BHR', 'BGD', 'BRB', 'BLR', 'BEL', 'BLX', 'BLZ', 'BEN', 'BMU', 'BTN', 'BOL', 'BIH', 'BWA', 'BAT', 'BRA', 'IOT', 'VGB', 'BRN', 'BGR', 'BFA', 'BDI', 'KHM', 'CMR', 'CAN', 'CPV', 'CYM', 'CAF', 'TCD', 'CHL', 'CHN', 'CXR', 'CCK', 'COL', 'COM', 'ZAR', 'COG', 'COK', 'CRI', 'CIV', 'HRV', 'CUB', 'CYP', 'CZE', 'CSK', 'DNK', 'DJI', 'DMA', 'DOM', 'TMP', 'ECU', 'EGY', 'SLV', 'GNQ', 'ERI', 'EST', 'ETH', 'ETF', 'EUN', 'FRO', 'FLK', 'FJI', 'FIN', 'PCZ', 'ZW1', 'TAN', 'VDR', 'SVR', 'ZPM', 'ATF', 'FRA', 'FRE', 'GUF', 'PYF', 'GAB', 'GMB', 'GAZ', 'GEO', 'DDR', 'DEU', 'GHA', 'GIB', 'GRC', 'GRL', 'GRD', 'GLP', 'GUM', 'GTM', 'GIN', 'GNB', 'GUY', 'HTI', 'VAT', 'HND', 'HKG', 'HUN', 'ISL', 'IND', 'IDN', 'IRN', 'IRQ', 'IRL', 'ISR', 'ITA', 'JAM', 'JPN', 'JTN', 'JOR', 'KAZ', 'KEN', 'KIR', 'PRK', 'KOR', 'KWT', 'KGZ', 'LAO', 'LVA', 'LBN', 'LSO', 'LBR', 'LBY', 'LIE', 'LTU', 'LUX', 'MAC', 'MKD', 'MDG', 'MWI', 'MYS', 'MDV', 'MLI', 'MLT', 'MHL', 'MTQ', 'MRT', 'MUS', 'MEX', 'FSM', 'MID', 'MDA', 'MCO', 'MNG', 'MSR', 'MAR', 'MOZ', 'MMR', 'NAM', 'NRU', 'NPL', 'NLD', 'ANT', 'NZE', 'NCL', 'NZL', 'NIC', 'NER', 'NGA', 'NIU', 'NFK', 'MNP', 'NOR', 'OMN', 'PCE', 'PAK', 'PLW', 'PAN', 'PNG', 'PRY', 'PMY', 'PER', 'PHL', 'PCN', 'POL', 'PRT', 'PRI', 'QAT', 'REU', 'ROM', 'RUS', 'RWA', 'RYU', 'SBH', 'SHN', 'KN1', 'SPM', 'WSM', 'SMR', 'STP', 'SWK', 'SAU', 'SEN', 'SYC', 'SLE', 'SIK', 'SGP', 'SVK', 'SVN', 'SLB', 'SOM', 'ZAF', 'SVU', 'ESP', 'SPE', 'LKA', 'KNA', 'LCA', 'VCT', 'SDN', 'SUR', 'SJM', 'SWZ', 'SWE', 'CHE', 'SYR', 'TWN', 'TJK', 'TZA', 'THA', 'TGO', 'TKL', 'TON', 'TTO', 'TUN', 'TUR', 'TKM', 'TCA', 'TUV', 'UGA', 'UKR', 'ARE', 'GBR', 'USA', 'UNS', 'URY', 'USP', 'UZB', 'VUT', 'VEN', 'VNM', 'VIR', 'WAK', 'WLF', 'ESH', 'YDR', 'YEM', 'SER', 'YUG', 'ZMB', 'ZWE']

def clean_df(df):
    df.index = df['Country Code']
    df = df[df['Country Code'].isin(country_list)]
    df = df.drop(columns=['Indicator Code', 'Country Code',
       '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
       '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989'])
    df = df.dropna(how='all')
    return df

fem_lit = clean_df(fem_lit)
male_lit = clean_df(male_lit)
```

```python
# Function for finding the max, min, and mean values of each dataset
def country_mean(df):
    df = df[['1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017']].T
    country_mean = pd.DataFrame(df.mean())
    country_mean = country_mean.rename(columns={0: 'Mean Literacy Rate'})
    return(country_mean)
#     df = pd.DataFrame(country_mean, country_max, country_min)
    
fem_lit_mean = country_mean(fem_lit)
male_lit_mean = country_mean(male_lit)
male_lit_mean
```

```python
lit_ratio = fem_lit_mean / male_lit_mean
# lit_ratio = lit_ratio.dropna()
lit_ratio = lit_ratio.rename(columns={"Mean Literacy Rate": "Ratio of Female to Male Literacy Rates"})
```

```python
shapefile = '../Resources/ne_10m_admin_0_countries/ne_10m_admin_0_countries.shp'

colors = 9
cmap = 'Blues'
figsize = (16, 10)
```

```python
gdf = gpd.read_file(shapefile)[['ADM0_A3', 'geometry']].to_crs('+proj=robin')
gdf.head()
```

```python
merged = gdf.merge(lit_ratio, left_on='ADM0_A3', right_on='Country Code')
merged = merged.merge(fem_lit_mean, left_on='ADM0_A3', right_on='Country Code')
merged = merged.merge(male_lit_mean, left_on='ADM0_A3', right_on='Country Code', suffixes=(' among Women', ' among Men'))
merged.sample(50)
```

```python
items=['Ratio of Female to Male Literacy Rates', 'Mean Literacy Rate among Women', 'Mean Literacy Rate among Men']
for item in items:
    ax = merged.dropna().plot(column=item, cmap=cmap, figsize=figsize, scheme='equal_interval', k=colors, legend=True)
    merged[merged[item].isna()].plot(ax=ax, color='#fafafa', hatch='///')
    ax.get_legend().set_bbox_to_anchor((.12, .4))
    ax.set_axis_off()
    ax.set_xlim([-1.5e7, 1.7e7])
    title = item
    ax.set_title(title, fontdict={'fontsize': 20}, loc='center')
    ax.figure.savefig(f'{title}.png')
#     images.append(imageio.imread(ax))
    
```
